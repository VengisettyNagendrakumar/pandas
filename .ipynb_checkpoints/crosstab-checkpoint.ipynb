```python
import pandas as pd
df=pd.read_excel('data.xlsx')
df
```

```python
pd.crosstab(df.gender,df.hobbies)
```

```python
pd.crosstab(df.gender,df.hobbies,margins=True)
```

```python
pd.crosstab(df.gender,[df.hobbies,df.age],margins=True)
```

```python
import numpy as np
pd.crosstab([df.gender],[df.hobbies],values=df.age,aggfunc=np.average)
```

```python
pd.crosstab(df.gender,df.hobbies,values=df.age,aggfunc=np.average)
```

```python
pd.crosstab([df.gender],[df.hobbies],normalize='index')
```
