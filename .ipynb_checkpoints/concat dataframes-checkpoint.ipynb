```python
import pandas as pd
india_weather=pd.DataFrame({
    'city':['mumbai','delhi','gujarat'],
    'temp':[32,34,56],
    'humidity':[80,34,56]
})
us_weather=pd.DataFrame({
    'city':['newyork','chicago','sans'],
    'temp':[89,87,78],
    'humidity':[23,13,54]
})
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
d=pd.concat([india_weather,us_weather],keys=['india','us'])
d
```

```python
d.loc['india']
```

```python
d.loc['us']
```

```python
temp_df=pd.DataFrame({
    'city':['mumbai','delhi','hand'],
    'temp':[34,56,78]
})
wind_df=pd.DataFrame({
    'city':['mumbai','delhi','hand'],
    'wind_speed':[5,65,87]
})
df=pd.concat([temp_df,wind_df],axis=1)
df
```

```python
wf=pd.DataFrame({
    'city':['mumbai','delhi'],
    'wind_speed':[5,87]
},index=[1,0])
wf
```

```python
temp_df=pd.DataFrame({
    'city':['mumbai','delhi','hand'],
    'temp':[34,56,78]
})
s=pd.Series(['humid','dry','wet'],name="event")
s
```

```python
d=pd.concat([temp_df,s],axis=1)
d
```
