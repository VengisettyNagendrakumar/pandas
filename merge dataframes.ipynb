```python
import pandas as pd
df1=pd.DataFrame({
    'city':['newyork','ornaldo','chicago'],
    'temp':[23,4,5]
})
df2=pd.DataFrame({
    'city':['newyork','chicago','ornaldo'],
    'humidity':[56,34,56]
})
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df1=pd.DataFrame({
    'city':['newyork','ornaldo','chicago','delhi'],
    'temp':[23,4,5,56]
})
df2=pd.DataFrame({
    'city':['newyork','chicago','ornaldo','sanfran'],
    'humidity':[56,34,56,45]
})
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```
